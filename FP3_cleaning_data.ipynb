# Importing and reading the dataset from Github link

```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/GioOng/DS3000_finalProject/main/bank-full.csv", delimiter=';')
data
```

# Creating a dictionary with the appropriate numerical replacement for each string value

```python
name_rep_dict = {'unknown': 0, 'management': 1, 'technician': 2, 'entrepreneur': 3, 'blue-collar': 4, 'retired': 5,
                'admin.': 6, 'servies': 7, 'self-employed': 8, 'unemployed': 9, 'housemaid': 10, 'student': 11,
                'single': 0, 'married': 1, 'divorced': 2, 'primary': 1, 'secondary': 2, 'tertiary': 3, 'no': 0, 'yes': 1,
                'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6, 'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10,
                'nov': 11, 'dec':12, 'success': 1, 'failure': 2,  'other': 3}
```

# Replacing all string categorical values into their corresponding numerical values

```python
data = data.drop(columns=['day', 'contact'])
data.replace(name_rep_dict, inplace=True)
data
```
